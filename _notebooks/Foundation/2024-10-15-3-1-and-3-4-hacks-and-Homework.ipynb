```python
#3.4

# Popcorn Hack 1: Create and Print Three Strings

# Creating three strings
greeting = "Hello, Jowan!"
hobby = "I love playing video games."
aspiration = "I plan to study game design in college."

# Printing the strings
print(greeting)
print(hobby)
print(aspiration)

# Popcorn Hack 2: String Concatenation with Your Own Name

# String concatenation with name and more details
first_name = "Jowan"
last_name = "Elzein"
interest = "video games"
future_plan = "game design"

print("Your name is: " + first_name + " " + last_name + ".")
print("You love " + interest + " and plan to study " + future_plan + ".")

# Hack 2: Function to Count Vowels in a String

def count_vowels(input_string):
    vowels = "aeiouAEIOU"  
    count = 0

    # Loop through each character in the string
    for char in input_string:
        if char in vowels:
            count += 1

    return count

string_to_check = input("Enter a string: ")
vowel_count = count_vowels(string_to_check)
print("Number of vowels in the string:", vowel_count)

# Hack 3: Function to Replace 'abc' with 'xyz'

def replace_abc_with_xyz(input_string):
    updated_string = input_string.replace('abc', 'xyz')
    return updated_string

string_to_modify = input("Enter a string: ")
modified_string = replace_abc_with_xyz(string_to_modify)
print("Modified string:", modified_string)

# Hack 4: Function to Check if a String is a Palindrome

import string

def is_palindrome(input_string):
    # Remove non-alphanumeric characters and convert to lowercase
    cleaned_string = ''.join(char.lower() for char in input_string if char.isalnum())
    # Check if the cleaned string is the same forwards and backwards
    return cleaned_string == cleaned_string[::-1]

string_to_check = input("Enter a string: ")
if is_palindrome(string_to_check):
    print("The string is a palindrome!")
else:
    print("The string is not a palindrome.")

# Hack 5: Function to Find the Longest Word in a Sentence

def find_longest_word(sentence):
    words = sentence.split()
    longest_word = ""
    
    for word in words:
        if len(word) > len(longest_word):
            longest_word = word
    
    return longest_word

sentence = input("Enter a sentence: ")
longest_word = find_longest_word(sentence)
print("The longest word in the sentence is:", longest_word)
```
