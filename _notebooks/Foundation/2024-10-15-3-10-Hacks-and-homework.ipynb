```python
from functools import reduce
import random

# Python Hack 1: List Management

def display_list(aList):
    print("Current list:", aList)

def add_item(aList, item):
    aList.append(item)
    print(f"'{item}' added to the list.")

def remove_item(aList, item):
    if item in aList:
        aList.remove(item)
        print(f"'{item}' removed from the list.")
    else:
        print(f"'{item}' not found in the list.")

def check_item(aList, item):
    if item in aList:
        print(f"'{item}' is in the list.")
    else:
        print(f"'{item}' is not in the list.")

def find_max(aList):
    if not aList:
        return None
    max_val = aList[0]
    for num in aList:
        if num > max_val:
            max_val = num
    return max_val

def find_min(aList):
    if not aList:
        return None
    min_val = aList[0]
    for num in aList:
        if num < min_val:
            min_val = num
    return min_val

def list_manager():
    user_list = []
    while True:
        print("\nChoose an operation: add, remove, check, display, max, min, quit")
        operation = input("Enter operation: ").lower()

        if operation == 'add':
            item = input("Enter an item to add: ")
            # Try to convert to int if possible, else keep as string
            try:
                item = int(item)  # Change to float(item) for floats
            except ValueError:
                pass  # Keep it as a string
            add_item(user_list, item)

        elif operation == 'remove':
            item = input("Enter an item to remove: ")
            remove_item(user_list, item)

        elif operation == 'check':
            item = input("Enter an item to check: ")
            check_item(user_list, item)

        elif operation == 'display':
            display_list(user_list)

        elif operation == 'max':
            max_val = find_max(user_list)
            if max_val is not None:
                print("Max value in the list:", max_val)
            else:
                print("The list is empty.")

        elif operation == 'min':
            min_val = find_min(user_list)
            if min_val is not None:
                print("Min value in the list:", min_val)
            else:
                print("The list is empty.")

        elif operation == 'quit':
            print("Exiting program.")
            break

        else:
            print("Invalid operation. Please try again.")

# Call the list manager function
list_manager()

# JavaScript Hack: Summing a List of Numbers
# List of numbers
numbers = [1, 2, 3, 4, 5]

# Using reduce to calculate the sum
sum_of_numbers = reduce(lambda x, y: x + y, numbers)
print("Sum of numbers:", sum_of_numbers)

# Second hack: random integers manipulation
# List of integers
random_integers = [3, 4, 8, 9, 1, 80, 77]

# Shuffle the list
random.shuffle(random_integers)
print("Random order:", random_integers)

# Sort the list
sorted_integers = sorted(random_integers)
print("Sorted integers:", sorted_integers)

# Shopping Cart Functionality
def display_cart(cart):
    print("\nShopping Cart:", cart)

def add_to_cart(cart, item):
    cart.append(item)
    print(f"'{item}' has been added to the cart.")

def remove_from_cart(cart, item):
    if item in cart:
        cart.remove(item)
        print(f"'{item}' has been removed from the cart.")
    else:
        print(f"'{item}' is not in the cart.")

def shopping_cart():
    cart = []
    while True:
        print("\nChoose an action: add, remove, display, quit")
        action = input("Enter action: ").lower()

        if action == 'add':
            item = input("Enter an item to add: ")
            add_to_cart(cart, item)

        elif action == 'remove':
            item = input("Enter an item to remove: ")
            remove_from_cart(cart, item)

        elif action == 'display':
            display_cart(cart)

        elif action == 'quit':
            print("Exiting the shopping cart program.")
            break

        else:
            print("Invalid action. Please try again.")

# Call the shopping cart function
shopping_cart()
```
